```python
from rdkit import Chem
from rdkit.Chem import Draw
import random
import pandas as pd
```

```python
# Sample 5 random SMILES strings from rl_1.csv
csv_path = '../runs/2026-02-11_14-05-32/Stage_4_RL/rl_1.csv'

# sample 5 random incides from the csv file

df = pd.read_csv(csv_path)
random_indices = random.sample(range(len(df)), 20)
smiles_list = df['SMILES'].iloc[random_indices].tolist()
mols = [Chem.MolFromSmiles(smi) for smi in smiles_list]
img_grid = Draw.MolsToGridImage(mols, molsPerRow=5, subImgSize=(200,200), legends=smiles_list)
img_grid
```
